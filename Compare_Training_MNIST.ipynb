# Compare Training

ResNet-18/34 variant trained on MNIST:

- weight decay applied to all trainable parameters

- weight decay _not_ applied to batch-norm and biases

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load training without weight decay applied to batch-norm and biases-
with open("ConvResNet_plLARS_MNIST_train_history.pkl", "rb") as file:
    train_history = pickle.load(file)

# Load training with weight decay-
with open("ConvResNet_LARS_wtdecay_MNIST_train_history.pkl", "rb") as file:
    train_history_wt_decay = pickle.load(file)
del file
```

```python
plt.figure(figsize = (12, 10))
plt.plot(list(train_history.keys()), [train_history[k]['acc'] for k in train_history.keys()], label = 'train acc')
plt.plot(list(train_history_wt_decay.keys()), [train_history_wt_decay[k]['acc'] for k in train_history_wt_decay.keys()], label = 'train acc wt decay')
plt.plot(list(train_history.keys()), [train_history[k]['test_acc'] for k in train_history.keys()], label = 'test acc')
plt.plot(list(train_history_wt_decay.keys()), [train_history_wt_decay[k]['test_acc'] for k in train_history_wt_decay.keys()], label = 'test acc wt decay')
plt.title("ConvResNet: Accuracy")
plt.xlabel("epochs")
plt.ylabel("accuracy (%)")
plt.legend(loc = 'best')
plt.show()
```

```python
plt.figure(figsize = (12, 10))
plt.plot(list(train_history.keys()), [train_history[k]['loss'] for k in train_history.keys()], label = 'train loss')
plt.plot(list(train_history_wt_decay.keys()), [train_history_wt_decay[k]['loss'] for k in train_history_wt_decay.keys()], label = 'train loss wt decay')
plt.plot(list(train_history.keys()), [train_history[k]['test_loss'] for k in train_history.keys()], label = 'test loss')
plt.plot(list(train_history_wt_decay.keys()), [train_history_wt_decay[k]['test_loss'] for k in train_history_wt_decay.keys()], label = 'test loss wt decay')
plt.title("ConvResNet: Accuracy")
plt.xlabel("epochs")
plt.ylabel("loss)")
plt.legend(loc = 'best')
plt.show()
```

```python
plt.figure(figsize = (12, 10))
plt.plot(list(train_history.keys()), [train_history[k]['lr'] for k in train_history.keys()], label = 'no wt decay')
plt.plot(list(train_history_wt_decay.keys()), [train_history_wt_decay[k]['lr'] for k in train_history_wt_decay.keys()], label = 'wt decay')
plt.xlabel("epochs")
plt.ylabel("lr")
plt.title("ConvResNet: Learning-Rate")
plt.legend(loc = 'best')
plt.show()
```
